```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("data/Algerian_forest_fires_dataset_UPDATE.csv", header = 1)
```

```python
dataset.head()
```

```python
dataset.info()
```

## data Cleaning

```python
## missing value
dataset[dataset.isnull().any(axis=1)] ## if ihave to find any missing value or null value in the heading if its there there will be written True
# dataset.isnull().sum()
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
# converting the type of region into int
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove the null values
df = df.dropna().reset_index(drop = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## remove the 122nd row
df =df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in column name
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Chnage the required columns as integer data type

```python
df.columns
```

```python
df[["month","day","year","Temperature","RH","Ws"]] = df[["month","day","year","Temperature","RH","Ws"]].astype(int)
```

```python
df.info()
```

### Chnging the other columns to float datatype


```python
objects = [features for features in df.columns if df[features].dtypes=="O"] # shows which of the data type has object class 
```

```python
for i in objects:
    if i != 'Classes':
        df[i] =df[i].astype(float)
```

```python
df.info()
```

```python
## lets saved the cleaned dataset
df.to_csv("data/Algerian_forest_fires_dataset_cleaned.csv",index = False) # index = False will help csv file not to save the index column
```
